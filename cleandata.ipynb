```python
import pandas as pd
import numpy as np

import datetime
from dateutil.parser import parse
```

# gather and clean data

```python
dt = pd.read_csv("twitter-sentiment-analysis.csv", header=None)
```

```python
print(dt.shape)
dt.head()
```

```python
dt = dt.rename(columns={0:"target", 1:"id", 2:"datetime", 3:"NO_QUERY", 4:"username", 5:"text"})
dt
```

```python
print(dt.iloc[:, 0].unique())
print(dt[dt.iloc[:, 0]==0].shape[0], dt[dt.iloc[:, 0]==4].shape[0])
```

only have binary target: 0(negative), 4(positive)

```python
arr = np.array(dt.iloc[:, 2])
print(arr.shape)
arr[:10]
```

```python
def str_datetime(s):
    ss = parse(s).strftime('%a %Y-%m-%d %H:%M:%S')
    return ss[:3], ss[4:]
```

```python
weekday, datetime = str_datetime(arr[0])
weekday, datetime
```

```python
tmp = dt.iloc[:, 2].apply(str_datetime)
```

```python
tmp
```

```python
week_date_time = pd.DataFrame(list(tmp), columns=["weekday", "datetime"])
```

```python
week_date_time
```

```python
weekdaydummies = pd.get_dummies(week_date_time['weekday'], prefix='weekday', dtype=float)
weekdaydummies = pd.DataFrame(weekdaydummies, columns=['weekday_'+w for w in ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]])
```

```python
weekdaydummies_datetime = pd.concat([weekdaydummies, week_date_time['datetime']], axis=1)
```

```python
weekdaydummies_datetime
```

```python
weekdaydummies_datetime.iloc[:, :-1].apply(sum)
```
